```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline

DATA_FILE = 'data/crime_homicide_subset_modified.csv'

df = pd.read_csv(DATA_FILE)
```

```python
custom_style = {'axes.labelcolor': 'white',
                'xtick.color': 'white',
                'ytick.color': 'white',
                'axes.facecolor' :  'black',
                'figure.facecolor' : 'black',
                'font.size' : 16,
                'axes.titlesize' : 16,
                'axes.labelsize' : 16,
                'grid.color' : 'grey',
                'text.color' : 'white'}
sns.set_style("darkgrid", rc=custom_style)
```

# Number of crimes by hour


```python
g = sns.factorplot(x='hour', data=df, kind='count', size=8)
g.set_axis_labels('Hour', 'Number of Crimes')
```

# Number of crimes by hour - sex abuse only.
All murders are reported at midnight

```python
g = sns.factorplot(x='hour', data=df[df.OFFENSE == "SEX ABUSE"], kind='count', size=8)
g.set_axis_labels('Hour', 'Number of Crimes')
```

# Number of crimes by weekday

```python
g = sns.factorplot(x='weekdaynb', data=df, kind='count', size=8)
g.set_axis_labels('Weekday', 'Number of Crimes')
g.set(xticklabels=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
```

# Number of crimes by weekday and offense

```python
g = sns.factorplot(x='weekdaynb', data=df, hue="OFFENSE", kind='count', size=8, palette="Reds")
g.set_axis_labels('Weekday', 'Number of Crimes')
g.set(xticklabels=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
```

# Sexual abuse by week hour 

```python
g = sns.factorplot(x='weekhour', data=df[df.OFFENSE == "SEX ABUSE"], kind='count', size=8)
g.set_axis_labels('Week hour', 'Number of Crimes')

labels = []
for i in xrange(0,168):
    labels.append('')
labels[0*24 + 12] = "Monday"
labels[1*24 + 12] = "Tuesday"
labels[2*24 + 12] = "Wednesday"
labels[3*24 + 12] = "Thursday"
labels[4*24 + 12] = "Friday"
labels[5*24 + 12] = "Saturday"
labels[6*24 + 12] = "Sunday"

g.set(xticklabels=labels)
```
